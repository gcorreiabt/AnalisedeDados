PCA (Principal Component Analysis), ou Análise de Componentes Principais em português, é um algoritmo de redução de dimensionalidade que transforma um conjunto de dados em um novo conjunto de dimensões menores, chamados de componentes principais. O objetivo do PCA é reduzir a complexidade do conjunto de dados, mantendo a maior parte da variação dos dados.

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# Carregando o iris dataset
iris = load_iris()
X = iris.data
```

```python
# Criando uma instância do PCA
pca = PCA(n_components=2)

# Aplicando o PCA nos dados

X_pca = pca.fit_transform(X)
```

```python
# Exibindo o novo formato dos dados
print(X_pca.shape)
```

```python
# Exibindo as duas primeiras componentes principais
print(pca.components_)
```

```python
# Exibindo a variação explicada por cada componente principal
print(pca.explained_variance_ratio_)
```

```python
# Plotando o gráfico de dispersão dos dados com duas dimensões principais
plt.scatter(X_pca[:,0], X_pca[:,1], c=iris.target, cmap='viridis')
plt.xlabel('Componente principal 1')
plt.ylabel('Componente principal 2')
plt.show()
```

Nesse exemplo, o dataset Iris é carregado utilizando a função load_iris do Scikit-learn, e os dados são armazenados na variável X. Em seguida, uma instância do PCA é criada utilizando a classe PCA, e o parâmetro n_components é definido como 2 para indicar que queremos reduzir a dimensionalidade para duas dimensões. O método fit_transform é utilizado para aplicar o PCA aos dados e gerar o novo conjunto de dados em duas dimensões, armazenado na variável X_pca. Em seguida, o novo formato dos dados é exibido utilizando o método shape. Além disso, as duas primeiras componentes principais são exibidas utilizando a propriedade components_, e a variação explicada por cada componente principal é exibida utilizando a propriedade explained_variance_ratio_.

Nesse exemplo, utilizamos o método scatter do Matplotlib para plotar o novo conjunto de dados X_pca em um gráfico de dispersão, onde cada ponto é colorido de acordo com a classe dos dados no dataset Iris (armazenado na propriedade target). As dimensões principais são exibidas no eixo x e no eixo y. A opção cmap='viridis' define a paleta de cores utilizada para as classes.

# Análise de componentes Principais (PCA)

A análise de componentes principais é uma técnica usada para identificação de um número menor de variáveis não correlacionadas conhecidas como componentes principais de um conjunto maior de dados.

A técnica é amplamente usada para enfatizar a variação e capturar padrões fortes em um conjunto de dados. A PCA é considerada um método estatístico útil e usado em campos como compressão de imagens, reconhecimento facial, neurociência e computação gráfica.

# O que é PCA

É um método de obtenção de variáveis importantes (na forma de componentes) a partir de um grande conjunto de variáveis disponíveis em um conjunto de dados.

# Como o PCA faz isso? 

Ele extrai um conjunto de recursos de baixa dimensão (componentes principais), fazendo uma projeção de dimensões irrelevantes de um conjunto de dados de alta dimensão (muitas variáveis) com o objetivo de capturar o máximo de informações possível

Com menos variáveis obtidas e, ao mesmo tempo, minimizando a perda de informações, a visualização também se torna muito mais significativa.

O PCA é mais útil ao lidar com dados tridimensionais ou superiores, ou seja 3 ou mais variáveis.

# Requisitos prévios:

. Variáveis devem ser numéricas e ter dados padronizados. Os componentes principais são fornecidos com a versão normalizada dos preditores originais. Isso ocorre porque os preditores originais podem ter escalas diferentes.

## Por exemplo:

Imagine um conjunto de dados com unidadesde medida de variáveis como galões, quilômetros, anos-luz, etc. É certo que a escala de varincias nessas variáveis será grande.
Executar PCA em variáveis não normalizadas levará a carregamentos insanamente grandes para variáveis com alta variância.

. A necessidade de redução de dimensionalidade:

> Dimensões grandes são difíceis de treinar, precisam de mais poder computacional e tempo.
> A visualização não é possível com dados dimensionais muito grandes.
> Carregar dados dimensionais muito altos pode ser um problema com espaço de armazenamento limitado na memória.
> Ele pode ser usado para reduzir a dimensão dos recursos, potencialmente levando a um melhor desempenho para o algoritmo de aprendizagem, removendo recursos redundantes, obsoletos e altamente correlacionados.

# Imagine isso:

Imagine, você está trabalhando em um projeto de ciência de dados em grand escala. O que acontece quando o conjunto de dados fornecidos tem muitas variáveis? Aqui estão algumas situações possíveis que você pode encontrar:

1. Você descobre que a maioria das variáveis são correlacionadas na análise.
2. Você perde a paciência e decide executar um modelo com todos os dados. Isso retorna uma precisão pobre e você se sente péssimo.
3. Você fica indeciso sobre o que fazer.
4. Você começa a pensar em algum método estratégico para encontrar algumas variáveis importantes.
   
O PCA Ajuda a superar essas dificuldades. Extraindo um conjonto de variáveis de um grande conjunto de variáveis existentes. Essas variáveis recém-extraídas são chamadas de Componentes Principais.

. Um componente principal é uma combinação linear das variáveis originais;
. Os componentes principais são extraídos de tal forma que o primeiro componente principal explica a variação máxima no conjunto de dados;
. O segundo componente principal tenta explicar a variação restante no conjunto de dados e não está correlacionado com primeiro componente principal;
. O terceiro componente principal tenta explicar a variação que não é explicada pelos dois primeiros componentes principais e assim por diante.

Cada dimensão adicional que adicionamos à técnica PCA captura cada vez menos a variação do modelo. O primeiro componente é o mais importante.

Por meio do PCA, você consegue extrair os fatores mais importantes dos dados.

```python
# Problema: Precisamos reduzir a dimensionalidade dos dados.

## Importando as bibliotecas
import numpy as np
from sklearn.decomposition import PCA
```

```python
# Importando o dataset iris
from sklearn.datasets import load_iris
data = load_iris()
```

```python
# Cria o Array e faz a preparação das variáveis de entrada X
X = iris.data
X
```

```python
# Cria a máquina preditiva de redução de dimensionalidade
pca = PCA(n_components=2)
pca.fit(X)
```

```python
# Avalia o resultado da máquina
print(pca.explained_variance_ratio_) # O valor explica qual a porcentagem de explicação de cada componente
print(pca.singular_values_) # Valores do componentes principais
```
