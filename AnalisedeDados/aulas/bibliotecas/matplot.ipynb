```python
import matplotlib.pyplot as plt
```

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

desprezo = [1.4, 0.7, 1.4, 1.2, 0.5, 1.0, 0.5, 1.2, 0.7, 0.7, 0.9, 1.9]
produtividade = [50.7, 58.7, 62.2, 62.3, 66.0, 59.5, 45.2, 59.3, 69.9, 63.1, 77.1, 76.6]
fidelidade = [32.6, 34.0, 40.3, 39.2, 33.3, 37.8, 39.5, 38.9, 39.7, 32.4, 29.4, 39.1]

meta_desp = []

for i in meses:
  meta_desp.append(0.8)

meta_prod = []

for i in meses:
  meta_prod.append(50)

meta_fid = []

for i in meses:
  meta_fid.append(34)
```

```python
plt.style.use("bmh")
plt.title('Indicador de desprezo de bolsa (2024)', size="20", color="r")
plt.xlabel('Ms')
plt.ylabel('% de bolsas desprezadas')
plt.plot(meses, desprezo, color= '#1231ed', marker = '.', label= 'Bolsas desprezadas')
plt.plot(meta_desp, dash_capstyle='projecting', label= 'Meta')
plt.legend()

plt.show()
```

```python
figura = plt.figure(figsize=(15, 6))
figura.suptitle('Apresentao de indicadores 2024', color="r")


figura.add_subplot(231)

plt.title('Indicador de desprezo')
plt.plot(meses, desprezo, color= '#1231ed', marker = '.', label= '% de Desprezo')
plt.plot(meta_desp, dash_capstyle='projecting', label= 'Meta')
plt.legend()
plt.xlabel('Ms')
plt.ylabel('% de bolsas desprezadas')

figura.add_subplot(232)

plt.title('Indicador de produtividade')
plt.plot(meses, produtividade, color= '#f23a9d', marker = '.', label= '% de Produtividade')
plt.plot(meta_prod, dash_capstyle='projecting', label= 'Meta')
plt.legend()
plt.xlabel('Ms')
plt.ylabel('% de produtividade')

figura.add_subplot(233)

plt.title('Indicadores de fidelidade')
plt.plot(meses, fidelidade, color= '#846dea', marker = '.', label= '% de Fidelizao')
plt.plot(meta_fid, dash_capstyle='projecting', label= 'Meta')
plt.legend()
plt.xlabel('Ms')
plt.ylabel('% de fidelidade')

plt.show()
```

```python
plt.scatter(meses, desprezo, marker='.')
plt.show()
```

```python
barras = plt.figure(figsize=(50,5))

barras.add_subplot(131)
plt.style.use('ggplot')
plt.bar(meses, desprezo, color='red', label= desprezo)

plt.show()
```

#Estilos para o grfico

##Linestyle or ls (Estilo de linha)
| Simbolo | Descrio |
| ----- | ------ |
| '-' | Linha reta |
| ':' | Linha pontilhada |
| '--' | Linhas retas com espaamento |
| '-.' | Linhas e pontos de forma revezada |
| (0,(0.01,2)) | Linha retas com espaamento, mas de forma personalizvel |

As linhas podem ter sua cor alterada utilizando color= 'cor'
As linhas podem sofrer alterao de tamanho utilizando linewidth = '2'
Ex. plt.plot(x, y, ls = ':', color = 'red', linewidth= '2')

<br>

##capstyle or dash_capstyle (estilo de borda de linha)
| Simbolo | Descrio |
| ----- | ------ |
| 'butt' | Linha padro, termina com um trao reto e quadrado |
| 'round' | Linha com as suas bordas arredondadas |
| 'projecting' | Similar ao padro butt, mas este passa um pouco do limite da linha |

Ex. plt.plot(x, y, capstyle = 'round')

## Ttulos e legendas

Podemos adicionar um **Ttulo ao grfico** utilizando plt.title('Ttulo do grfico'), alm disso, podemos realizar algumas alteraes nesse comando, como alterar a cor e tamanho da fonte.

Ex. plt.title('Indicadores (2024)', size="20", color="r")

Para adicionar **Legendas para as linhas** do grficos, podemos acrescentar dentro do cdigo plot, o comando label='Nome', dessa forma, criaremos uma legenda para as linhas utilizadas. Em seguida, apresentamos estas descries criadas utilizando o comando legend()

Ex:
plt.plot(x, y, marker='.', ls='--', label='Nome da linha/descrio')
plt.legend()

Para adicionar os **Ttulos da linha X e Y** no grfico, podemos utilizar os comando plt.xlabel('nome') e plt.ylabel('nome'), dessa forma,  possvel identificar as informaes do grfico.

Ex:
plt.xlabel('Meses do  ano')
plt.ylabel('Salrio')

## Figure

O mtodo plt.figure em matplotlib  usado para criar uma nova janela de figura, que  onde o grfico  desenhado. Isso  til quando voc deseja criar vrios grficos em uma nica janela ou quando deseja personalizar a aparncia da janela da figura (por exemplo, definir o tamanho da janela).

Este mtodo pode receber cdigos de personalizao, como por exemplo:

|   Cdigo   |   Descrio   |
| ------------- | ---------------- |
| plt.figsize(largura, altura) | Comando utilizado inicialmente para definir o tamanho da figura, definindo o tamanho e largura da figura. Este comando  usado armazenando a informao dentro de uma varivel. |
| var.suptitle('titulo') | Comando usado para acrescentar um ttulo a figura geral |
| var.add_subplot() | Este comando  usado para definir as caracteristicas do grfico que ser acrescentado na figura. O primeiro valor define a quantidade de linhas necessaria, o segundo a quantidade de colunas e o terceiro a posio que esse grfico ira ocupar dentro da figura. |

## Outros grficos

plt.scatter() : pode ser usado da mesma forma que o comando plot, mas o scatter  usado para criar um grfico de pontos, diferente do plot, que gera grficos de linhas.

plt.bar() : possui um conceito de criao semelhante aos grficos anteriores, mas este no possui alguns cdigos como marker, capstyle e linestyle.

```python
# DESAFIO

# Voc  um analista de dados de uma empresa de alimentos e recebeu a tarefa de analisar as vendas
# de um novo produto em diferentes regies do Brasil. Voc precisa criar um grfico que mostre a quantidade de vendas do
# produto em cada regio, de forma clara e objetiva. Use a biblioteca Matplotlib do Python para criar um grfico de barras que mostre a quantidade de vendas em cada regio.

# Dados
# regioes = ['Norte', 'Nordeste', 'Centro-Oeste', 'Sudeste', 'Sul']
# vendas = [1500, 2500, 1800, 3000, 2700]



regioes = ['Norte', 'Nordeste', 'Centro-Oeste', 'Sudeste', 'Sul']
vendas = [1500, 2500, 1800, 3000, 2700]

plt.title("Vendas dos produtos por regio")

plt.bar(regioes, vendas, label='Venda por regio R$')
plt.xlabel('Regio')
plt.ylabel('Vendas')


plt.legend()
plt.show()
```
