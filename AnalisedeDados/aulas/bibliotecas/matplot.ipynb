```python
import matplotlib.pyplot as plt
```

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

desprezo = [1.4, 0.7, 1.4, 1.2, 0.5, 1.0, 0.5, 1.2, 0.7, 0.7, 0.9, 1.9]
produtividade = [50.7, 58.7, 62.2, 62.3, 66.0, 59.5, 45.2, 59.3, 69.9, 63.1, 77.1, 76.6]
fidelidade = [32.6, 34.0, 40.3, 39.2, 33.3, 37.8, 39.5, 38.9, 39.7, 32.4, 29.4, 39.1]

meta_desp = []

for i in meses:
  meta_desp.append(0.8)

meta_prod = []

for i in meses:
  meta_prod.append(50)

meta_fid = []

for i in meses:
  meta_fid.append(34)
```

```python
plt.style.use("bmh")
plt.title('Indicador de desprezo de bolsa (2024)', size="20", color="r")
plt.xlabel('Mês')
plt.ylabel('% de bolsas desprezadas')
plt.plot(meses, desprezo, color= '#1231ed', marker = '.', label= 'Bolsas desprezadas')
plt.plot(meta_desp, dash_capstyle='projecting', label= 'Meta')
plt.legend()

plt.show()
```

```python
figura = plt.figure(figsize=(15, 6))
figura.suptitle('Apresentação de indicadores 2024', color="r")


figura.add_subplot(231)

plt.title('Indicador de desprezo')
plt.plot(meses, desprezo, color= '#1231ed', marker = '.', label= '% de Desprezo')
plt.plot(meta_desp, dash_capstyle='projecting', label= 'Meta')
plt.legend()
plt.xlabel('Mês')
plt.ylabel('% de bolsas desprezadas')

figura.add_subplot(232)

plt.title('Indicador de produtividade')
plt.plot(meses, produtividade, color= '#f23a9d', marker = '.', label= '% de Produtividade')
plt.plot(meta_prod, dash_capstyle='projecting', label= 'Meta')
plt.legend()
plt.xlabel('Mês')
plt.ylabel('% de produtividade')

figura.add_subplot(233)

plt.title('Indicadores de fidelidade')
plt.plot(meses, fidelidade, color= '#846dea', marker = '.', label= '% de Fidelização')
plt.plot(meta_fid, dash_capstyle='projecting', label= 'Meta')
plt.legend()
plt.xlabel('Mês')
plt.ylabel('% de fidelidade')

plt.show()
```

```python
plt.scatter(meses, desprezo, marker='.')
plt.show()
```

```python
barras = plt.figure(figsize=(50,5))

barras.add_subplot(131)
plt.style.use('ggplot')
plt.bar(meses, desprezo, color='red', label= desprezo)

plt.show()
```

#Estilos para o gráfico

##Linestyle or ls (Estilo de linha)
| Simbolo | Descrição |
| ----- | ------ |
| '-' | Linha reta |
| ':' | Linha pontilhada |
| '--' | Linhas retas com espaçamento |
| '-.' | Linhas e pontos de forma revezada |
| (0,(0.01,2)) | Linha retas com espaçamento, mas de forma personalizável |

As linhas podem ter sua cor alterada utilizando color= 'cor'
As linhas podem sofrer alteração de tamanho utilizando linewidth = '2'
Ex. plt.plot(x, y, ls = ':', color = 'red', linewidth= '2')

<br>

##capstyle or dash_capstyle (estilo de borda de linha)
| Simbolo | Descrição |
| ----- | ------ |
| 'butt' | Linha padrão, termina com um traço reto e quadrado |
| 'round' | Linha com as suas bordas arredondadas |
| 'projecting' | Similar ao padrão butt, mas este passa um pouco do limite da linha |

Ex. plt.plot(x, y, capstyle = 'round')

## Títulos e legendas

Podemos adicionar um **Título ao gráfico** utilizando plt.title('Título do gráfico'), além disso, podemos realizar algumas alterações nesse comando, como alterar a cor e tamanho da fonte.

Ex. plt.title('Indicadores (2024)', size="20", color="r")

Para adicionar **Legendas para as linhas** do gráficos, podemos acrescentar dentro do código plot, o comando label='Nome', dessa forma, criaremos uma legenda para as linhas utilizadas. Em seguida, apresentamos estas descrições criadas utilizando o comando legend()

Ex:
plt.plot(x, y, marker='.', ls='--', label='Nome da linha/descrição')
plt.legend()

Para adicionar os **Títulos da linha X e Y** no gráfico, podemos utilizar os comando plt.xlabel('nome') e plt.ylabel('nome'), dessa forma, é possível identificar as informações do gráfico.

Ex:
plt.xlabel('Meses do  ano')
plt.ylabel('Salário')

## Figure

O método plt.figure em matplotlib é usado para criar uma nova janela de figura, que é onde o gráfico é desenhado. Isso é útil quando você deseja criar vários gráficos em uma única janela ou quando deseja personalizar a aparência da janela da figura (por exemplo, definir o tamanho da janela).

Este método pode receber códigos de personalização, como por exemplo:

|   Código   |   Descrição   |
| ------------- | ---------------- |
| plt.figsize(largura, altura) | Comando utilizado inicialmente para definir o tamanho da figura, definindo o tamanho e largura da figura. Este comando é usado armazenando a informação dentro de uma variável. |
| var.suptitle('titulo') | Comando usado para acrescentar um título a figura geral |
| var.add_subplot() | Este comando é usado para definir as caracteristicas do gráfico que será acrescentado na figura. O primeiro valor define a quantidade de linhas necessaria, o segundo a quantidade de colunas e o terceiro a posição que esse gráfico ira ocupar dentro da figura. |

<br>

## Outros gráficos

plt.scatter() : pode ser usado da mesma forma que o comando plot, mas o scatter é usado para criar um gráfico de pontos, diferente do plot, que gera gráficos de linhas.

plt.bar() : possui um conceito de criação semelhante aos gráficos anteriores, mas este não possui alguns códigos como marker, capstyle e linestyle.

```python
# DESAFIO

# Você é um analista de dados de uma empresa de alimentos e recebeu a tarefa de analisar as vendas
# de um novo produto em diferentes regiões do Brasil. Você precisa criar um gráfico que mostre a quantidade de vendas do
# produto em cada região, de forma clara e objetiva. Use a biblioteca Matplotlib do Python para criar um gráfico de barras que mostre a quantidade de vendas em cada região.

# Dados
# regioes = ['Norte', 'Nordeste', 'Centro-Oeste', 'Sudeste', 'Sul']
# vendas = [1500, 2500, 1800, 3000, 2700]



regioes = ['Norte', 'Nordeste', 'Centro-Oeste', 'Sudeste', 'Sul']
vendas = [1500, 2500, 1800, 3000, 2700]

plt.title("Vendas dos produtos por região")

plt.bar(regioes, vendas, label='Venda por região R$')
plt.xlabel('Região')
plt.ylabel('Vendas')


plt.legend()
plt.show()
```
