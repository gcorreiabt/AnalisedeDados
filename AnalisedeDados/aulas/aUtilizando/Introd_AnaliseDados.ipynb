# **Desafio**
## Descrição

Você é um cientista de dados e precisa analisar a relação entre as variáveis altura e peso em uma amostra de pessoas. Escreva um programa contendo as informações de altura e peso de cada pessoa, calcule a correlação entre as duas variáveis e exiba um gráfico de dispersão com as duas variáveis.

## Dados

pesos = [60.2, 73.5, 64.0, 68.1, 87.0, 75.4, 70.0, 76.8, 81.2, 58.9, 69.7, 92.5, 78.6, 66.3, 71.8, 79.5, 63.7, 72.5, 83.0, 76.1]

alturas = [1.65, 1.75, 1.68, 1.72, 1.90, 1.80, 1.77, 1.82, 1.87, 1.60, 1.74, 2.00, 1.78, 1.67, 1.73, 1.83, 1.62, 1.76, 1.90, 1.79]

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pesos = [60.2, 73.5, 64.0, 68.1, 87.0, 75.4, 70.0, 76.8, 81.2, 58.9, 69.7, 92.5, 78.6, 66.3, 71.8, 79.5, 63.7, 72.5, 83.0, 76.1]
alturas = [1.65, 1.75, 1.68, 1.72, 1.90, 1.80, 1.77, 1.82, 1.87, 1.60, 1.74, 2.00, 1.78, 1.67, 1.73, 1.83, 1.62, 1.76, 1.90, 1.79]

correlacao = np.corrcoef(pesos, alturas)[0,1]

plt.style.use('bmh')

plt.scatter(pesos, alturas, color= 'r', marker= '+', label= 'Altura X Peso')

plt.title(f'Correlação {correlacao:.2f}')
plt.xlabel('Peso')
plt.ylabel('Altura')
plt.legend()

print(correlacao)
```
