O algoritmo k-means é um algoritmo de agrupamento (clustering) não supervisionado amplamente utilizado em machine learning. Ele agrupa os dados em k clusters, onde k é um número pré-definido pelo usuário. O algoritmo k-means é um algoritmo iterativo que começa com k centroides aleatórios, atribui cada ponto de dados ao cluster cujo centróide está mais próximo e, em seguida, recalcula o centróide de cada cluster. Esse processo é repetido até que a diferença entre os centróides atuais e os anteriores seja inferior a um valor definido.

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Carregando o dataset Iris
iris = load_iris()
X = iris.data

# Definindo o número de clusters
k = 3

# Criando uma instância do algoritmos K-means
kmeans = KMeans(n_clusters=k)

# Treinando o Modelo
kmeans.fit(X)

# Exibindo os centroides dos Clusters
print(kmeans.cluster_centers_)

# Exibindo as previsões de cluster para os dados
print(kmeans.labels_)

# Plotando os clusters e os centróides correspondentes
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1], marker='x', color='r')

plt.show()
```

Nesse exemplo, o dataset Iris é carregado utilizando a função load_iris do Scikit-learn, e os dados são armazenados na variável X. Em seguida, o número de clusters é definido como 3 e uma instância do algoritmo k-means é criada utilizando a classe KMeans. O modelo é treinado utilizando o método fit, e os centróides dos clusters são exibidos utilizando a propriedade cluster_centers_. Por fim, as previsões de cluster para os dados são exibidas utilizando a propriedade labels_.

Lembrando que o algoritmo k-means utiliza todos os atributos do dataset, nesse exemplo são utilizados 4 atributos, todavia, ao plotar o gráfico só é possível mostrar duas dimensões, dessa forma ocorre esses casos que 'aparentam' estar misturados, mas se for analisado as 4 dimensões isso não ocorre.

Utilizando o método do cotovelo (Elbow method) para determinar o melhor valor de K 

```python
# Percorrendo diferentes valores de K
valores_k = []
inercias = []

for i in range (1,15):
    kmeansi = KMeans(n_clusters=i, random_state=0).fit(X)   # Analisar o espaçamento entre os valores, quanto menor o espaço entre um valor e outro, menor o valor da clusterização
    valores_k.append(i)
    inercias.append(kmeansi.inertia_)
```

```python
# visualizando a relação entre K e inercia
fig, ax = plt.subplots()

ax.plot(valores_k, inercias)

plt.show()
```

```python
#visualizando os targets e os labels no mesmo gráfico
fig, ax = plt.subplots(ncols=2, figsize=(20, 7))

ax[0].scatter(X['petal length (cm)'], X['petal width (cm)'], c=kmeans.labels_)
ax[1].scatter(X['petal length (cm)'], X['petal width (cm)'], c=y)

plt.show()
```

# Usando indice rand para avaliar os acertos (pois sabemos os rótulos)

. mede a semelhança entre 2 pontos
. Depende de termos e rótulos / labels

onde:
. a: numero de pares que pertencem a mesma classe e ao mesmo cluster
. b: numero de pares que pertencem a classes diferentes e a cluster diferentes
. c: numero de pares que pertencem a mesma classe e cluster diferentes
. d: numero de pares que pertencem a classes diferentes e ao mesmo cluster

> o indice rand tem alguns problemas como ter valores altos mesmo para dados aleatórios e aumentar a medida que aumentamos o número de grupos. Para resolver, utilizamos o Indice Rand Ajustado, onde:

. para uma clusterização aleatória, seu valor é 0
. O máximo é 1
. Valores negativos representam clusterizações piores que escolher de forma aleatória os clusters

```python
# importando o rand (metrics)
from sklearn import metrics
```

```python
# verificando o adjusted_rand_score para K = 2
metrics.adjusted_rand_score(y, kmeans.labels_)
```

```python
# verificando o adjusted_rand_score para K = 3
metrics.adjusted_rand_score(y, kmeans3.labels_)
```

```python
# verificando o adjusted_rand_score para K = 4
metrics.adjusted_rand_score(y, kmeans4.labels_)
```

```python
# podemos percorrer os valores de k calculando esses dois indicadores
valores_k = []
ARI = []
RI = []

for i in range (1,15):
    kmeansi = KMeans(n_clusters=i, random_state=0).fit(X)   # Analisar o espaçamento entre os valores, quanto menor o espaço entre um valor e outro, menor o valor da clusterização
    valores_k.append(i)
    RI.append(metrics.rand_score(y, kmeansi.labels_))
    ARI.append(metrics.adjusted_rand_score(y, kmeansi.labels_))
```

```python
fig, ax = plt.subplots()

ax.plot(valores_k, ARI)
ax.plot(valores_k, RI)

plt.show()
```

# Coeficiente silhueta para avaliação dos dados (Melhor na opinião do professor)

. Quanto maior esse coeficiente, mais bem definidos são os clusters do modelo
. vamos considerar 2 pontuações
> a: A distância média entre uma amostra e todos os outros pontos da mesma classe.
> b: A distância média entre uma amostra e todos os outros pontos no outro cluster mais próximo.

. Ele avalia tanto a distância intracluster (O quanto os pontos estão afastados dentro do próprio cluster) quanto a distância interclusters (O quanto os clusters estão afastados entre si)

```python
# Verificando o silhouette_score para k=2
metrics.silhouette_score(X, kmeans.labels_)
```

```python
# Verificando o silhouette_score para k=3
metrics.silhouette_score(X, kmeans3.labels_)
```

```python
valores_k = []
s = []


for i in range (2,15):  #Para o silhouette_score, é necessário no mínimo 2 cluster para o cálculo, com somente 1, apresenta erro
    kmeansi = KMeans(n_clusters=i, random_state=0).fit(X)   # Analisar o espaçamento entre os valores, quanto menor o espaço entre um valor e outro, menor o valor da clusterização
    valores_k.append(i)
    s.append(metrics.silhouette_score(X, kmeansi.labels_))
```

```python
fig, ax = plt.subplots()

ax.plot(valores_k, s)

plt.show()
```

Pelas métricas, o melhor resultado foi obtido fazendo K = 2. Então esse é o melhor valor de K? Não exatamante
. Isso é apenas um direcionamento, porém o que é mais importante é conhecermos o negócio!
